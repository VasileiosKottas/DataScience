Entropy and Relative Entropy - distance Kullback-Leiber

```python
import numpy as np
from math import log2
import matplotlib.pyplot as plt
```

```python
#%% Entropy
# Calculation of p: nparray
def calc_entropy(p):
    H = 0
    for i in range(0, p.size):
        if p[i] == 0: continue
        H += p[i]*log2(1/p[i])
    return H
# Calculate relative entropy
def calc_relative_entropy(p, q):
    D = 0
    for i in range(0, p.size):
        if q[i] == 0: continue
        D += p[i]*log2(p[i]/q[i])
    return D
```

```python
#%% Throwing a coin
P_coin = np.arange(start=0, stop=1.1, step=0.1)
H_arr = np.array([])
for p_head in P_coin:
    p_tail=1-p_head
    p = np.array([p_head, p_tail])
    H_arr = np.append(H_arr, calc_entropy(p))
plt.plot(P_coin, H_arr, 'b-o')
plt.grid()
plt.show()
```

```python
#%% Max entropy of n-categories
N = np.arange(start=2, stop=17, step=1)
H_max = np.array([])
for n in N:
    p = np.ones(n)/n
    H_max = np.append(H_max, calc_entropy(p))
plt.plot(N, H_max, 'b-o')
plt.grid()
plt.show()
```

```python
#%% Relative Entrpy, distance Kullback-Leiber
p = np.array([1/3, 1/3, 1/3])
q = np.array([2/3, 1/6, 1/6])
calc_relative_entropy(p, q)
```
